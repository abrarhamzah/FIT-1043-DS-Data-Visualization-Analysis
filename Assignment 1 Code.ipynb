```python
dataframe = pd.read_csv('CityPairs.csv', usecols=["AustralianPort", "Passengers_Total", "Freight_Total_(tonnes)", "Mail_Total_(tonnes)","Year"],)
dataframe
```

```python
#here I use groupby() function to aggregate the data to get the sum of passengers each year
data_group = dataframe.groupby(['AustralianPort','Year']).sum()
data_group
```
